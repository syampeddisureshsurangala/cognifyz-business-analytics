```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/Data_set 2.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
savings_objectives_column = df['What are your savings objectives?']

# Display the unique savings objectives
print("Unique Savings Objectives:")
print(savings_objectives_column.unique())

# Count the occurrences of each savings objective
savings_objectives_counts = savings_objectives_column.value_counts()

# Display the counts
print("\nSavings Objectives Distribution:")
print(savings_objectives_counts)

plt.figure(figsize=(10, 6))
savings_objectives_counts.plot(kind='bar', color='lightblue')
plt.title('Savings Objectives Distribution')
plt.xlabel('Savings Objective')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
