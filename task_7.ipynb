```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/Data_set 2.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
info_sources_column = df['Source']

# Display the unique information sources
print("Unique Information Sources:")
print(info_sources_column.unique())

# Count the occurrences of each information source
info_sources_counts = info_sources_column.value_counts()

# Display the counts
print("\nInformation Sources Distribution:")
print(info_sources_counts)

plt.figure(figsize=(10, 6))
info_sources_counts.plot(kind='bar', color='lightcoral')
plt.title('Information Sources Distribution')
plt.xlabel('Information Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
info_sources_column = df['Source'].dropna()

# Display the unique information sources
print("Unique Information Sources:")
print(info_sources_column.unique())

# Count the occurrences of each information source
info_sources_counts = info_sources_column.value_counts()

# Display the counts
print("\nInformation Sources Distribution:")
print(info_sources_counts)
```
