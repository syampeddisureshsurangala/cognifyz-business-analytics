```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/Data_set 2.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
# Identify the investment column (assuming it's named 'Investment_Avenue')
investment_column = df['Mutual_Funds']

# Display the unique investment avenues
print("Unique Investment Avenues:")
print(investment_column.unique())

# Count the occurrences of each investment avenue
investment_counts = investment_column.value_counts()

# Display the counts
print("\nInvestment Avenue Distribution:")
print(investment_counts)

plt.figure(figsize=(10, 6))
sns.countplot(x=investment_column)
plt.title('Distribution of Investment Avenues')
plt.xlabel('Investment Avenue')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
most_frequent_investment = investment_counts.idxmax()
most_frequent_count = investment_counts.max()

print(f"\nThe investment avenue with the highest frequency is '{most_frequent_investment}' with {most_frequent_count} occurrences.")
```
